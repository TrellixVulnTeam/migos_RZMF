```python
culture = culture.replace('[','')
culture = culture.replace(']', '')

culture_ar = culture.split('\n\n')

chorus = list(filter(lambda x: 'Chorus:' in x, culture_ar))[0]

r = re.compile('Verse [0-9]')
verses = [i for i in culture_ar if r.match(i)]
```

```python
#df creator
prompts = []
completions = [] 
for i in range(0,len(verses)):
    line = verses[i]
    splits = line.split('\n')
    first_line = splits[0]
    completion = '\n'.join(splits[1:])

    others = [x for x in verses if x != line]
    prompt = '\n\n'.join(others)

    prompt = prompt + first_line
    completion = ' ' + completion

    prompts.append(prompt)
    completions.append(completion)    
```

```python
def clean_data(s):
    s = s.replace('[','')
    s = s.replace(']', '')

    s_ar = s.split('\n\n')

    chorus = list(filter(lambda x: 'Chorus:' in x, s_ar))[0]

    r = re.compile('Verse [0-9]')
    verses = [i for i in s_ar if r.match(i)]

    return(verses)
```

```python
prompts = []
completions = []
for key, value in all_songs.items():
    verses = clean_data(value)

    for i in range(0,len(verses)):
        line = verses[i]
        splits = line.split('\n')
        first_line = splits[0]
        completion = '\n'.join(splits[1:])

        others = [x for x in verses if x != line]
        prompt = '\n\n'.join(others)

        prompt = 'Title: ' + key + '\n' + prompt + first_line
        completion = ' ' + completion

        prompts.append(prompt)
        completions.append(completion)
```
